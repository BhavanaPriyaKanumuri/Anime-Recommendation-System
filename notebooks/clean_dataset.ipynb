## Import pandas, read the dataset

```python
import pandas as pd
```

```python
anime_data = pd.read_csv('DATA/Anime_data.csv')
```

## Camel-case the columns

```python
anime_data.columns = map(str.lower, anime_data.columns)
anime_data
```

## Clean up the columns

```python
import re
def clean_up(s):
    return re.sub(r"[^\sa-zA-Z0-9\.\,\;\!\-]","",str(s))
```

```python
anime_data['title'] = anime_data['title'].apply(clean_up)
anime_data['genre'] = anime_data['genre'].apply(clean_up)
anime_data['producer'] = anime_data['producer'].apply(clean_up)
anime_data['studio'] = anime_data['studio'].apply(clean_up)
```

## Selecting columns from dataset

```python
anime_data = anime_data[['title','genre', 'synopsis','producer','studio','rating','scoredby','members', 'source','aired']]
anime_data
```

## Dropping NaN's

```python
anime_data = anime_data.dropna(axis=0)
anime_data
```

## Changing 'scoredby' and 'members' from float values to integer values

```python
anime_data = anime_data.astype({'scoredby':'int64','members':'int64'})
anime_data
```

#### Removing rows with "nan" values and Hentai genre

```python
anime_data = anime_data[~anime_data.genre.str.contains("Hentai", na=False)]
anime_data = anime_data[~anime_data.title.str.contains("Hentai", na=False)]
anime_data = anime_data[~anime_data.producer.str.contains("nan", na=False)]
anime_data = anime_data[~anime_data.studio.str.contains("nan", na=False)]
```

```python
anime_data
```

## Further refining anime features to have 'scoredby' greater than 1000

```python
anime_data =  anime_data[anime_data['scoredby']>=1000]
anime_data
```

```python
anime_data.title.nunique()
```

```python
anime_data = anime_data.drop_duplicates(subset='title', keep="first")
```

```python
anime_data = anime_data.reset_index()
anime_data = anime_data.drop(columns='index')
anime_data. index = pd.RangeIndex(start=1, stop=3186, step=1)
anime_data
```

```python
anime_data.to_csv("DATA/anime_cleaned.csv")
```
